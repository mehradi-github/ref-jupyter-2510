```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(np.random.rand(50,4),columns=['a','b','c','d'])
df
```

```python
k=df.plot.scatter(x='a',y='b',color='red', label='Group1')
df.plot.scatter(x='c',y='d',color='blue', label='Group2',ax=k)
```
